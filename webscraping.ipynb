# 2 Read .csv from MUltiple URLs with Pandas

- https://www.football-data.co.uk/mmz4281/2122/E0.csv
- https://www.football-data.co.uk/mmz4281/2122/E1.csv
- https://www.football-data.co.uk/mmz4281/2122/E2.csv
- https://www.football-data.co.uk/mmz4281/2223/E3.csv
- https://www.football-data.co.uk/mmz4281/2122/EC.csv

##### Link: root + season + league

```python
# link structure
"https://www.football-data.co.uk/mmz4281/" + "2122" + "/" + "E0" + ".csv"
```

```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

```python
# creating list od leagues
leagues = ['E1', 'E1', 'E2', 'E3', 'EC']

frames = []

# looping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + '2122' + '/' + league + '.csv', encoding='unicode_escape')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show !st, 2nd, 3rd element
frames[0]
```

```python
frames[4]
```

# 2.2 Multiple Seasons

```python
for season in range(15, 21):
    print(str(season) + str(season+ 1))
```

```python
# creating list od leagues
leagues = ['E1', 'E2', 'E3']

frames = []

# looping through leagues, read multiple csv and append it into a list
for league in leagues:
    for season in range(15, 21):
        df = pd.read_csv(root + (str(season) + str(season+ 1)) + '/' + league + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# total frames (6 seasons * 3 leagues)
len(frames)
```
